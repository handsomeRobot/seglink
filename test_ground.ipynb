```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from preprocessing.ssd_vgg_preprocessing import rotate_image_with_bounding_box
```

```python
def plt_show_image(image, title):
    plt.figure(figsize=(20, 20))
    plt.imshow(image)
    plt.title(title)
    plt.show()
    
def draw_polybox(img, bboxes, show=False, title=''):
    
    '''
    :param img: input image, np.ndarray
    :param bboxes: list of bbox
    '''
    
    img = cv2.polylines(img.copy(), bboxes, True, [255, 0, 0], 5) 
    if show:
        plt_show_image(img, title)
    return img
```

```python
image = "/home/kxu/workspace/data/kuaidi/shentong20171110/00015.jpg"

image = cv2.imread(image)
print(image.shape)
print(image)
plt_show_image(image, title="input")


bboxes = np.array([[900, 500], [1200, 500], [1200, 1800], [900, 1800]])
draw_polybox(image, [bboxes], show=True)
print(bboxes.shape)

xs = bboxes[:, 0].reshape([-1])
ys = bboxes[:, 1].reshape([-1])
print(xs, ys)
image, bboxes, xs, ys = rotate_image_with_bounding_box(image, bboxes, xs, ys)
```
